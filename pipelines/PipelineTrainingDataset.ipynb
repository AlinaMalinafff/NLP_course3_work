```python
from docx import Document
import re
import pymorphy2
import stanza
from collections import defaultdict
from gensim.models import KeyedVectors
import random
import firebase_admin
from firebase_admin import credentials, db
from google.colab import files
import os
from tqdm import tqdm
import shutil
from typing import List, Dict, Optional, Tuple
import json

# Инициализация анализаторов
morph = pymorphy2.MorphAnalyzer()
stanza_pipeline = stanza.Pipeline(lang='ru', processors='tokenize,pos,lemma,depparse')

class SmartAugmenter:
    def __init__(self, model_path, semantic_threshold=0.6, max_synonyms=5):
        self.model = KeyedVectors.load_word2vec_format(model_path, binary=False)
        self.cache = defaultdict(list)
        self.semantic_threshold = semantic_threshold
        self.max_synonyms = max_synonyms

        self.manual_synonyms = {
            "процесс": ["работа", "деятельность", "ход", "продвижение"],
            "анализ": ["изучение", "исследование", "разбор", "оценка"],
            "отчёт": ["документ", "презентация", "отчетность", "доклад"]
        }

        self.blacklist = {"егэ", "впн", "интуитивизм"}

    def augment_sentence(self, text: str, target_changes: int = 3) -> str:
        tokens = re.findall(r'\w+|[^\w\s]', text, re.UNICODE)
        changes = 0
        candidates = []

        for i, token in enumerate(tokens):
            if not token.isalpha() or len(token) < 3:
                continue

            parsed = morph.parse(token.lower())[0]
            if parsed.tag.POS in {'PREP', 'CONJ', 'PRCL', 'NPRO', 'INTJ'}:
                continue

            lemma = parsed.normal_form
            candidates.append((i, token, lemma, parsed))

        random.shuffle(candidates)

        for i, orig_word, lemma, parsed in candidates:
            if changes >= target_changes:
                break

            synonym = self._get_best_synonym(lemma, parsed)
            if not synonym or synonym.lower() == orig_word.lower():
                continue

            inflected = self._inflect_word(synonym, parsed)
            if not inflected:
                continue

            if orig_word[0].isupper():
                inflected = inflected.capitalize()

            tokens[i] = inflected
            changes += 1

        return self._reconstruct_text(tokens)

    def _get_best_synonym(self, lemma, parsed):
        manual = self._get_manual_synonyms(lemma)
        if manual:
            return random.choice(manual)

        tag = parsed.tag
        pos_map = {
            'NOUN': '_NOUN', 'VERB': '_VERB', 'INFN': '_VERB',
            'ADJF': '_ADJ', 'ADJS': '_ADJ', 'ADVB': '_ADV'
        }
        pos_suffix = pos_map.get(tag.POS)
        if not pos_suffix:
            return None

        key = f"{lemma}{pos_suffix}"
        if key not in self.cache:
            try:
                similar = self.model.most_similar(key, topn=20)
                filtered = [
                    s[0].split('_')[0] for s in similar
                    if '_' in s[0] and s[0].endswith(pos_suffix) and s[1] >= self.semantic_threshold
                ]
                self.cache[key] = filtered[:self.max_synonyms]
            except KeyError:
                self.cache[key] = []

        synonyms = [w for w in self.cache[key] if w not in self.blacklist and w != lemma]
        return random.choice(synonyms) if synonyms else None

    def _get_manual_synonyms(self, word):
        for key, syns in self.manual_synonyms.items():
            if word == key:
                return [s for s in syns if s not in self.blacklist and s != word]
        return []

    def _inflect_word(self, word, parsed):
        try:
            new_parsed = morph.parse(word)[0]
            if 'Name' in parsed.tag or 'Surn' in parsed.tag:
                return word  # не склоняем имена
            inflected = new_parsed.inflect(parsed.tag.grammemes)
            return inflected.word if inflected else word
        except:
            return word

    def _reconstruct_text(self, tokens):
        text = ''
        for i, tok in enumerate(tokens):
            if i > 0 and tok not in ',.!?:;':
                text += ' '
            text += tok
        return text


class TextPreprocessor:
    """Расширенная предобработка текста"""

    @staticmethod
    def resolve_references(text: str) -> str:
        """Разрешение анафоры (замена местоимений существительными)"""
        words = text.split()
        last_noun = None
        for i, word in enumerate(words):
            parsed = morph.parse(word)[0]
            if 'NOUN' in parsed.tag:
                last_noun = word
            elif word.lower() in ["он", "она", "оно", "это"] and last_noun:
                words[i] = last_noun
        return ' '.join(words)

    @staticmethod
    def handle_ellipsis(text: str) -> str:
        """Объединение однородных сказуемых и восстановление эллипсиса"""
        if " и " in text and any(w in text for w in ["решили", "сделали", "завершили"]):
            return re.sub(r" и ", ", ", text)
        return text

    @staticmethod
    def normalize_text(text: str) -> str:
        """Нормализация текста для анализа"""
        text = text.lower()
        text = re.sub(r'[^\w\s]', '', text)  # Удаление пунктуации
        return text

class LinguisticFeatures:
    """Расширенные лингвистические правила и словари"""

    # Лексические маркеры
    PROCESS_LEX = {
        "процесс", "анализ", "разработка", "пока", "ещё не", "в процессе",
        "обсуждение", "планирование", "исследование", "изучение", "подготовка",
        "развитие", "формирование", "создание", "тестирование", "внедрение"
    }

    RESULT_LEX = {
        "результат", "итог", "заверш", "сделано", "вывод", "достигнут",
        "получен", "готов", "окончательный", "итоговый", "финальный",
        "реализован", "выполнен", "закончен", "подведён", "окончен"
    }

    # Глагольные формы
    PROCESS_VERBS = {
        "анализировать", "разрабатывать", "тестировать", "обсуждать",
        "планировать", "исследовать", "изучать", "готовить", "создавать"
    }

    RESULT_VERBS = {
        "завершить", "реализовать", "выполнить", "закончить", "подвести",
        "достигнуть", "получить", "сделать", "внедрить", "представить"
    }

    # Временные маркеры
    TIME_MARKERS = {
        "process": {
            "в процессе", "в ходе", "в течение", "на стадии", "на этапе",
            "пока", "ещё", "уже", "в настоящее время", "сейчас"
        },
        "result": {
            "в итоге", "в результате", "в конечном счёте", "в завершение",
            "наконец", "после", "по окончании", "по завершении"
        }
    }

    # Конструкции сравнения
    COMPARISON_PATTERNS = [
        r"не только\s+.+?\s+, но и\s+.+?",
        r"как\s+.+?\s+, так и\s+.+?",
        r"с одной стороны\s+.+?,\s+с другой стороны\s+.+?"
    ]

    @classmethod
    def detect_time_references(cls, text: str) -> Tuple[int, int]:
        """Определение временных ссылок в тексте"""
        process, result = 0, 0
        for marker in cls.TIME_MARKERS["process"]:
            if marker in text:
                process += 1
        for marker in cls.TIME_MARKERS["result"]:
            if marker in text:
                result += 1
        return process, result

    @classmethod
    def check_verb_aspect(cls, verb: str) -> Optional[str]:
        """Определение вида глагола (совершенный/несовершенный)"""
        parsed = morph.parse(verb)[0]
        if 'VERB' in parsed.tag:
            if 'perf' in parsed.tag:
                return "result"
            elif 'impf' in parsed.tag:
                return "process"
        return None

    @classmethod
    def check_comparisons(cls, text: str) -> Optional[Dict[str, int]]:
        """Анализ сравнительных конструкций"""
        for pattern in cls.COMPARISON_PATTERNS:
            if re.search(pattern, text):
                parts = re.split(r"не только|, но и|как|, так и|с одной стороны|, с другой стороны", text)
                process = any(p in cls.PROCESS_LEX for p in parts)
                result = any(p in cls.RESULT_LEX for p in parts)
                return {"process": int(process), "result": int(result)}
        return None

class SyntaxSemanticsHelper:
    """Помощник для синтаксико-семантического анализа"""

    @staticmethod
    def evaluate_verb_context(word, sentence) -> Optional[str]:
        """Проверка контекста глагола на предмет валентности и объектной завершенности"""
        if word.upos != 'VERB':
            return None

        obj_present = False
        subject_present = False

        for w in sentence.words:
            if w.head == word.id:
                if w.deprel in ('obj', 'xcomp', 'obl'):
                    obj_present = True
                elif w.deprel == 'nsubj':
                    subject_present = True

        # Оценка по шаблону: если есть субъект и объект — результативность
        if subject_present and obj_present:
            return 'result'
        elif subject_present:
            return 'process'
        return None

    @staticmethod
    def detect_frame_pattern(word: str) -> Optional[str]:
        """Упрощённая проверка фреймового ожидания глагола"""
        imperfective_goals = {'анализировать', 'изучать', 'обсуждать', 'планировать'}
        perfective_outcomes = {'завершить', 'представить', 'достигнуть', 'выполнить'}

        lemma = morph.parse(word)[0].normal_form
        if lemma in imperfective_goals:
            return 'process'
        elif lemma in perfective_outcomes:
            return 'result'
        return None

class HybridAnalyzer:
    """Улучшенный гибридный анализатор с расширенными правилами"""

    CONFIDENCE_THRESHOLD = 2
    LONG_SENTENCE_THRESHOLD = 15  # Количество слов в длинном предложении

    @classmethod
    def analyze(cls, text: str) -> str:
        """Основной метод анализа текста"""
        # Предобработка
        text = TextPreprocessor.resolve_references(text)
        text = TextPreprocessor.handle_ellipsis(text)
        normalized = TextPreprocessor.normalize_text(text)

        # Быстрый анализ
        markers = cls._rule_based_analysis(normalized, text)
        diff = abs(markers['process'] - markers['result'])

        # Глубокий анализ при неуверенности
        if diff < cls.CONFIDENCE_THRESHOLD:
            doc = stanza_pipeline(text)
            cls._syntax_analysis(doc, markers, normalized)
            cls._semantic_analysis(text, markers)

        # Определение результата с порогом уверенности
        if markers['process'] == markers['result']:
            return "Нейтральный"
        return "Процесс" if markers['process'] > markers['result'] else "Результат"

    @classmethod
    def _rule_based_analysis(cls, original_text, text: str) -> Dict[str, int]:
        """Правило-ориентированный анализ текста"""
        markers = defaultdict(int)

        # Токенизация и лемматизация
        lemmas = [morph.parse(word)[0].normal_form for word in text.split()]

        # 1. Лексические маркеры
        for lemma in lemmas:
            if lemma in LinguisticFeatures.PROCESS_LEX:
                markers['process'] += 1
            if lemma in LinguisticFeatures.RESULT_LEX:
                markers['result'] += 1

        # 2. Глагольные формы
        for lemma in lemmas:
            aspect = LinguisticFeatures.check_verb_aspect(lemma)
            if aspect:
                markers[aspect] += 1

        # 3. Временные маркеры (по исходному тексту, так как они устойчивые словосочетания)
        time_process, time_result = LinguisticFeatures.detect_time_references(text)
        markers['process'] += time_process
        markers['result'] += time_result

        # 4. Сравнительные конструкции
        comparison = LinguisticFeatures.check_comparisons(text)
        if comparison:
            markers['process'] += comparison['process']
            markers['result'] += comparison['result']

        # 5. Длина предложения
        if len(lemmas) > cls.LONG_SENTENCE_THRESHOLD:
            markers['process'] += 1
        elif len(lemmas) < 8:
            markers['result'] += 1

        # Контрастные конструкции
        contrast_markers = ['но', 'однако', 'хотя', 'несмотря на', 'в то же время']
        for marker in contrast_markers:
            if marker in original_text:
                parts = original_text.split(marker, 1)
                if len(parts) == 2:
                    second_part = parts[1]
                    for word in second_part.split():
                        lemma = morph.parse(word)[0].normal_form
                        if lemma in LinguisticFeatures.PROCESS_LEX:
                            markers['process'] += 1
                        if lemma in LinguisticFeatures.RESULT_LEX:
                            markers['result'] += 1


        return markers


    @classmethod
    def _syntax_analysis(cls, doc, markers: Dict[str, int], text: str):
        for sentence in doc.sentences:
            for word in sentence.words:
                # Вид глагола (дополнительная проверка)
                if word.upos == 'VERB':
                    parsed = morph.parse(word.text)[0]
                    if 'perf' in parsed.tag:
                        markers['result'] += 1
                    elif 'impf' in parsed.tag:
                        markers['process'] += 1

                    # Новый модульный контекстный анализ
                    context = SyntaxSemanticsHelper.evaluate_verb_context(word, sentence)
                    if context:
                        markers[context] += 1

                    # Фреймовая эвристика
                    frame = SyntaxSemanticsHelper.detect_frame_pattern(word.text)
                    if frame:
                        markers[frame] += 0.5

                # Сложные придаточные
                if word.lemma in ["хотя", "несмотря", "в то время как"]:
                    markers['process'] += 1
                elif word.lemma in ["поэтому", "следовательно", "таким образом"]:
                    markers['result'] += 1

                # Инфинитивные конструкции
                if word.feats and 'VerbForm=Inf' in word.feats:
                    markers['process'] += 0.5

    @classmethod
    def _semantic_analysis(cls, text: str, markers: Dict[str, int]):
        """Семантический анализ текста"""
        # Анализ модальности
        modal_words = ["должен", "должны", "нужно", "следует", "можем", "необходимо"]
        if any(word in text for word in modal_words):
            markers['process'] += 1

        intention_words = ["должен", "должны", "нужно", "следует", "планируем", "стоит", "предстоит"]
        if any(word in text for word in intention_words):
            markers['process'] += 1

        # Анализ отрицания
        if "не " in text or "нет " in text:
            markers['process'] += 0.5

        # Анализ вопросительных конструкций
        if "?" in text:
            markers['process'] += 1


# 1. Загрузка файлов в Colab
def upload_files():
    uploaded = files.upload()
    os.makedirs('input_docs', exist_ok=True)
    for filename in uploaded.keys():
        shutil.move(filename, f'input_docs/{filename}')
    return list(uploaded.keys())

# 2. Аугментация документов
def augment_documents(input_folder='input_docs', output_folder='augmented_docs', num_augmentations=1):
    os.makedirs(output_folder, exist_ok=True)
    augmenter = SmartAugmenter('ruscorpora_upos_skipgram_300_5_2018.vec')
    if not augmenter:
        print("Аугментатор не инициализирован, пропускаем аугментацию")
        return []

    augmented_files = []
    for filename in tqdm(os.listdir(input_folder), desc="Аугментация"):
        if filename.endswith('.docx'):
            doc = Document(f'{input_folder}/{filename}')
            for i in range(num_augmentations):
                aug_doc = Document()
                for para in doc.paragraphs:
                    if para.text.strip():
                        aug_text = augmenter.augment_sentence(para.text)
                        aug_doc.add_paragraph(aug_text)
                aug_filename = f'aug_{i}_{filename}'
                aug_doc.save(f'{output_folder}/{aug_filename}')
                augmented_files.append(aug_filename)
    return augmented_files

# 3. Парсинг и анализ документов для Realtime Database)))))))
def process_documents(db_ref, folders=['input_docs', 'augmented_docs']):
    analyzer = HybridAnalyzer()

    for folder in folders:
        for filename in tqdm(os.listdir(folder), desc=f"Обработка {folder}"):
            if filename.endswith('.docx'):
                doc = Document(f'{folder}/{filename}')
                full_text = "\n".join([p.text for p in doc.paragraphs if p.text.strip()])

                respondent_phrases = re.findall(
                    r"Респондент\s*:\s*(.*?)(?=\s*(?:Интервьюер|$))",
                    full_text,
                    flags=re.DOTALL
                )

                for phrase in respondent_phrases:
                    if not phrase.strip():
                        continue

                    label = analyzer.analyze(phrase.strip())

                    # Создаем новую запись в Realtime Database
                    new_entry_ref = db_ref.child('interview_data_try2').push()
                    new_entry_ref.set({
                        'text': phrase.strip(),
                        'label': label,
                        'source_file': filename,
                        'is_augmented': 'aug_' in filename,
                        'timestamp': {'.sv': 'timestamp'}
                    })

# 4. Инициализация Firebase для Realtime Database
def init_firebase():
    from google.colab import files

    # Проверяем наличие файла конфигурации
    if not any(f.endswith('.json') for f in os.listdir()):
        files.upload()

    # Проверяем, инициализировано ли приложение
    if not firebase_admin._apps:
        cred = credentials.Certificate('serviceAccountKey(2).json')
        firebase_admin.initialize_app(cred, {
            'databaseURL': 'https://nlp-coursework-try3-default-rtdb.europe-west1.firebasedatabase.app/'
        })
        print("Firebase Realtime Database успешно инициализирован!")
    else:
        print("Firebase уже инициализирован")

    # Получаем ссылку на корень базы данных
    return db.reference()

def analyze_paragraph_in_blocks(paragraph, n=3):
    doc = nlp(paragraph)
    sentences = [s.text.strip() for s in doc.sentences if s.text.strip()]
    blocks = [' '.join(sentences[i:i+n]) for i in range(0, len(sentences), n)]

    results = []
    for block in blocks:
        syntax_tree = nlp(block)
        labels = []

        for sentence in syntax_tree.sentences:
            for rule in rules:
                label = rule(sentence)
                if label:
                    labels.append(label)

        if labels:
            final_label = resolve_conflicts(labels)
            results.append({"text": block, "label": final_label})
        else:
            results.append({"text": block, "label": "NONE"})

    return results


# Основной пайплайн
def main():
    print("1. Загрузка файлов...")
    uploaded_files = upload_files()
    print(f"Загружено {len(uploaded_files)} файлов")

    print("\n2. Аугментация данных...")
    augmented_files = augment_documents(num_augmentations=1)
    print(f"Создано {len(augmented_files)} аугментированных файлов")

    print("\n3. Подключение к Firebase Realtime Database...")
    db_ref = init_firebase()

    print("\n4. Обработка и загрузка данных...")
    process_documents(db_ref)

    print("\nГотово! Все данные загружены в Firebase Realtime Database")
    print(f"Итого обработано: {len(uploaded_files)} исходных + {len(augmented_files)} аугментированных файлов")

if __name__ == "__main__":
    main()
```
